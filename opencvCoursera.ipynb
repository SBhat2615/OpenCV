# Image manipulation - 1 pixel at a time

OPENCV - Crop image, change pixel value

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread(image)
cropImg = img[100:400, 100:400, :]

rgbCropImg = cv2.cvtColor(cropImg, cv2.COLOR_BGR2RGB)

plt.imshow(rgbCropImg)
```

PIL - Crop image, change pixel value

```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
from PIL import ImageOps

img = Image.open(image)

left = 100
upper = 100
right = 400
lower = 400

cropImg = img.crop((left, upper, right, lower))

plt.imshow(cropImg)
```
